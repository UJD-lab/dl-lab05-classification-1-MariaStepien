## Temat:

Konwolucyjne sieci neuronowe w zagadnieniu klasyfikacji obrazów I

### Cel ćwiczenia:

Celem ćwiczenia jest wytrenowanie i przetestowanie modelu konwolucyjnej sieci neuronowej do klasyfikacji obrazów kwiatów.

### Wykorzystane narzędzia:

Visual Studio Code, Python z bibliotekami tensorflow, matplotlib, numpy, gradio, sklearn, seaborn, pathlib


### Zadanie 1


Model się nie przeuczył, fine-tuning rozpoczął się po 5 epoce i spowodował chwilowy spadek accuracy i wzrost straty. W kolejnych epokach model zaczął stopniowo wracać do wyniku przed fine tuning. W ostatniej epoce model osiągnął wyższy accuracy niż przed fine-tuning.

### Zadanie 2



Lepsze wyniki osiągnął eksperyment, w którym każda faza miała 10 epok. Eksperyment ten ukończył z lepszym accuracy oraz niższym loss. Pomimo zwiększenia ilości epok, model nie wykazuje przeuczenia.

### Zadanie 3

Najczęściej mylone są tulipany z różami. Model może mylić róże z tulipanami ze względu na kształt zamkniętej, nierozkwitniętej róży, który przypomina tulipana. Często są także mylone mniszki lekarskie z stokrotkami, co może być spowodowane podobnym kształtem płatków kwiatów.

### Zadanie 5

a. Zamrażanie bazy w fazie 1 ma na celu ochronę wcześniej wyuczonych wag, które model zdobył na zbiorze danych ImageNet.

b. Fine-tuning w fazie 2 pozwala na odmrożenie części wag bazy modelu i ich powolną aktualizację wraz z nowym klasyfikatorem. Dzięki temu model może lepiej dostosować wyuczone wcześniej uniwersalne cechy do specyficznych detali obecnych w konkretnym zbiorze danych.

c. Zastosowanie MobileNetV2 pozwala na uzyskanie szybszej zbieżności i zmniejszenie wymagań w stosunku do bazy danych (mniejsza ilość danych potrzebna do nauczenia modelu).

### Wnioski

Rozwiązanie zadań z laboratorium pozwoliło na zobaczenie jak zamrażanie bazy oraz fine-tuning pomagają w osiągnięciu wysokiego accuracy i niskiego loss, dodatkowo zwracając uwagę na to, że ilość epok ma duże znaczenie na otrzymywane wyniki.
